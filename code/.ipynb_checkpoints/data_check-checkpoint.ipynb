## Raw Data 확인
### Data : Log Data
--- 입증할만한 궁금증 (EDA)

- 대출 신청을 한 유저와 안한 유저 간의 퍼널 차이 / 
- 특정 view 체류시간 (e.g. 대출 신청 안하는 유저는 특히 한도조회 인트로에 체류 시간 적다.) / 
- 특정 view 확인 여부 (e.g. 대출 신청 하는 유저는 특히 한도조회 page 는 한 번은 들어가더라/)
- 방문당 event / 방문당 유니크 event 차이가 존재하는가 (e.g. 대출 하는 유저는 한 번 들어오면 4개 이상의 unique event 는 있더라.)
- 개별 경로 차이

--- 그냥 궁금증

- 어떤 이벤트 순서로 view 가 많을까 ? :
- 어떤 event(뷰)가 체류 시간이 길까? :
- 각 유저별 마지막 event 는 뭘까? : 
- 특히 view 가 많은 요일,시간이 있을까 ? :

```python
print("이벤트별_view_num")

log_action_cnt = log_data.groupby("행동명").유저_번호.count().reset_index(name='count').sort_values(['count'], ascending = False)
print(log_action_cnt)

print("이벤트별_view_bar_chart")
log_action_cnt.plot.barh(x="행동명", y='count', rot = 0, figsize=(15,5), fontsize = 15)
```
