````python
with open('../data/log_dict_user.pkl','rb') as f:
    log_dict_user = pickle.load(f)
````

````python
log_ngram_sort_df_15=pd.read_csv("../data/log_ngram_sort_df_15.csv")
log_ngram_sort_df_15.rename(columns={"Unnamed: 0":"route"}, inplace=True)
log_ngram_sort_df_15
````

````python
main_route_idx=[5, 6, 9, 10, 11, 12, 13, 14, 15, 16, 17, 19]
main_route_cols=['UG', 'GU', 'GV', 'UV', 'VG', 'UGV', 'GUV', 'VU', 'VUG', 'VGU', 'UVG', 'GVU']

funnel_route_idx=[7, 8, 3]
funnel_route_cols=['funnel_10', 'funnel_21', 'funnel_32']

````

````python
# log_dict_user[log_ngram_sort_df_15.iloc[main_route_idx[0]]['route']]
````

````python
agg_df[main_route_cols]=0
agg_df[funnel_route_cols]=0
````

````python
# testing

print(main_route_cols[0])
log_dict_by_user = log_dict_user[log_ngram_sort_df_15.iloc[main_route_idx[0]]['route']]
print("max: ", max(log_dict_by_user))

user_df = list(agg_df['유저_번호'])
dict_list=[0]*len(agg_df)
for i in tqdm(range(len(log_dict_by_user))):
    try:
        dict_list[user_df.index(log_dict_by_user[i])]+=1
    except ValueError: continue # 해당 유저가 유저 리스트에 없음       
agg_df[main_route_cols[0]]=dict_list
agg_df[main_route_cols[0]].describe()
````

````python
for i in range(len(main_route_cols)):
    log_dict_by_user = log_dict_user[log_ngram_sort_df_15.iloc[main_route_idx[i]]['route']]

    user_df = list(agg_df['유저_번호'])
    dict_list=[0]*len(agg_df)
    for j in tqdm(range(len(log_dict_by_user))):
        try:
            dict_list[user_df.index(log_dict_by_user[j])]+=1
        except ValueError: continue # 해당 유저가 유저 리스트에 없음       
    agg_df[main_route_cols[i]]=dict_list
agg_df[main_route_cols].describe()
````

````python
for i in range(len(funnel_route_cols)):
    log_dict_by_user = log_dict_user[log_ngram_sort_df_15.iloc[funnel_route_idx[i]]['route']]

    user_df = list(agg_df['유저_번호'])
    dict_list=[0]*len(agg_df)
    for j in tqdm(range(len(log_dict_by_user))):
        try:
            dict_list[user_df.index(log_dict_by_user[j])]+=1
        except ValueError: continue # 해당 유저가 유저 리스트에 없음       
    agg_df[funnel_route_cols[i]]=dict_list
agg_df[funnel_route_cols].describe()
````

````python
time_delta1=pd.read_csv("../data/time_delta1.csv")
time_delta2=pd.read_csv("../data/time_delta2.csv")
time_delta3=pd.read_csv("../data/time_delta3.csv")
````

````python
time_delta1.drop(['funnel_10_hesitate_x', 'funnel_10_hesitate_y'], axis=1, inplace=True)
time_delta2.drop(['funnel_21_hesitate_sec_x', 'funnel_21_hesitate_sec_y'], axis=1, inplace=True)
time_delta3.drop(
    ['funnel_32_hesitate_sec_x', 'funnel_32_hesitate_sec_y'], axis=1, inplace=True)
time_delta1.head()
````

````python
len(agg_df), len(time_delta1), len(time_delta2), len(time_delta3)
````

````python
agg_df.describe()
````

````python
agg_df=pd.merge(left = agg_df, right = time_delta1, how = "left", on = "유저_번호")
agg_df=pd.merge(left = agg_df, right = time_delta2, how = "left", on = "유저_번호")
agg_df=pd.merge(left = agg_df, right = time_delta3, how = "left", on = "유저_번호")
````
